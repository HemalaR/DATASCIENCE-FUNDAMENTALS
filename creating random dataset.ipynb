```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
#create a dataframe
#date_range is predefined keyword in pandas
ds=pd.DataFrame(np.random.rand(10,4),index=pd.date_range('1/1/2000',periods=10),columns=list('ABCD'))
```

```python
#ds local variable
#pd pandas variable
#x - numpy random range
#y - index year range -----specify the time period also i.e, 10 months
#columns - different lines in the graph below
ds.plot()
```

```python
sns.barplot(data=ds)
```

```python
ds.plot.bar()
```

```python
ds=pd.DataFrame(np.random.rand(10,4),columns=['A','B','C','D'])
```

```python
ds.plot.bar()
```

```python
#print/plot a stacked bar graph
ds=pd.DataFrame(np.random.rand(10,4),columns=['A','B','C','D'])
ds.plot.bar(stacked=True)
```

```python
#to plot bar graph horizontally
ds=pd.DataFrame(np.random.rand(10,4),columns=['A','B','C','D'])
ds.plot.barh(stacked=True)
```

```python
ds=pd.DataFrame(np.random.rand(10,4),columns=['A','B','C','D'])
ds.plot.box()
```

```python
#area plot
ds.plot.area()
```

```python
#scatterplot
#GIVE ARGUMENTS x and y for scatterplot
ds=pd.DataFrame(np.random.rand(40,4),columns=['A','B','C','D'])
ds.plot.scatter(x='A',y='B')
```
