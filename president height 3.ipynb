```python
sns.barplot(x='name',y='height(cm)',data=ds)
plt.ylim=(100,200)
plt.xticks(rotation=90)
plt.figure(figsize=(15,5))
plt.show()
```

```python
sns.barplot(x='name',y='height(cm)',data=ds,color='blue')
plt.title('president graph')
plt.xticks(rotation=90)
plt.xlim=(100,200)
plt.figure(figsize=(15,5))
plt.show()
```
