```python
a=[1,2,3,4,5,6,7,8,9]
India=[10,11,12,9,8,15,12,4,2]
Aus=[10,20,30,40,10,60,20,80,30]
plt.plot(a,India)
plt.plot(a,Aus)
plt.legend(["India","Aus"],loc="lower right")

plt.title("Runrate")
plt.xlabel("Over")
plt.ylabel("Runrate")
plt.show()
```
